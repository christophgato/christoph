```python
# Importar el módulo
from modulo import Punto2D
import math

# Puntos de prueba
A = Punto2D(math.pi, 6)
B = Punto2D(4, 1)

# Prueba de los atributos
print(f"A.x = {A.x}, A.y = {A.y}")  # Debe mostrar A.x = math.pi, A.y = 6

# Prueba de su representación
print(A)  # Debe mostrar (math.pi,6)

# Prueba de llamada como función de multiplicación por escalar
print(A(3))  # Debe mostrar (math.pi*3,18)

# Prueba de suma y resta de puntos, y multiplicación por escalar
C = 3 * (A + B)
D = A - B
print(f"C = {C}")  # Debe mostrar (3*(math.pi+4), 3*(6+1))
print(f"D = {D}")  # Debe mostrar (math.pi-4, 6-1)

# Prueba del gráfico
A.grafica()  # Debe mostrar el punto A en un gráfico

# Prueba de distancia entre dos puntos
print(f"Distancia entre A y B: {A.distancia(B)}")  # Debe calcular √((4-math.pi)² + (1-6)²)

# Prueba de distancia al origen
print(f"Distancia de A al origen: {abs(A)}")  # Debe mostrar el resultado de √((math.pi)² + 6²)
```
